```python
# ì…€ 1: ë°ì´í„° ë¡œë”© ë° ê¸°ë³¸ ì„¤ì •
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ import
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    confusion_matrix, accuracy_score, precision_score, 
    recall_score, roc_auc_score, f1_score,
    precision_recall_curve
)
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier  # VotingClassifier ì¶”ê°€
from sklearn.utils import resample
import psutil
import gc

# ë©”ëª¨ë¦¬ ì‚¬ìš©ëŸ‰ì„ ì¤„ì´ê¸° ìœ„í•œ ì„¤ì •
pd.options.mode.chained_assignment = None

# ë°ì´í„° ë¡œë“œ ì‹œ ë©”ëª¨ë¦¬ ìµœì í™”
def load_data(file_path):
    return pd.read_csv(file_path, low_memory=True)

# ë°ì´í„° ë¡œë“œ
base = load_data('Base.csv')
var1 = load_data('Variant I.csv')
var2 = load_data('Variant II.csv')
var3 = load_data('Variant III.csv')
var4 = load_data('Variant IV.csv')
var5 = load_data('Variant V.csv')

# ë°ì´í„°ì…‹ ë”•ì…”ë„ˆë¦¬ ìƒì„±
datasets = {
    'base': base,
    'var1': var1,
    'var2': var2,
    'var3': var3,
    'var4': var4,
    'var5': var5
}

# ë©”ëª¨ë¦¬ í•´ì œë¥¼ ìœ„í•œ í•¨ìˆ˜
def clear_memory():
    gc.collect()
```

###  1. ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸ ë° ë°ì´í„° ë¡œë”©

**ëª©ì **: í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ë¶ˆëŸ¬ì˜¤ê³  6ê°œì˜ ì‹ ìš©ì¹´ë“œ ì‚¬ê¸° íƒì§€ ë°ì´í„°ì…‹ì„ ë¡œë“œí•©ë‹ˆë‹¤.

**ì£¼ìš” êµ¬ì„±ìš”ì†Œ**:
- **ë°ì´í„° ì²˜ë¦¬**: pandas, numpy
- **ë¨¸ì‹ ëŸ¬ë‹ ëª¨ë¸**: XGBoost, LightGBM, RandomForest
- **í‰ê°€ ë©”íŠ¸ë¦­**: AUC, Precision, Recall, F1-score
- **ë©”ëª¨ë¦¬ ìµœì í™”**: `low_memory=True` ì˜µì…˜ìœ¼ë¡œ ëŒ€ìš©ëŸ‰ ë°ì´í„° íš¨ìœ¨ì  ì²˜ë¦¬

**ë°ì´í„°ì…‹ ì„¤ëª…**:
- Base: ê¸°ë³¸ ë°ì´í„°ì…‹ (100ë§Œ ê±´)
- Variant I~V: ê°ê¸° ë‹¤ë¥¸ íŠ¹ì„±ì„ ê°€ì§„ ë³€í˜• ë°ì´í„°ì…‹

 **íŒ**: MacBook Air M2ì˜ ë©”ëª¨ë¦¬ í•œê³„ë¥¼ ê³ ë ¤í•˜ì—¬ ìµœì í™”ëœ ë¡œë”© ë°©ì‹ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.

```python
# ì…€ 2: ë°ì´í„° ì „ì²˜ë¦¬ í•¨ìˆ˜ ì •ì˜
def preprocess_data(df):
    # ë¶ˆí•„ìš”í•œ ì»¬ëŸ¼ ì œê±°
    drop_col = [
        'payment_type', 'employment_status', 'prev_address_months_count',
        'intended_balcon_amount', 'housing_status', 'days_since_request'
    ]
    df = df.drop(columns=drop_col)
    
    # ì´ìƒì¹˜ ì²˜ë¦¬
    df = df[df['current_address_months_count'] >= 0]
    df['bank_months_count'] = df['bank_months_count'].replace({-1: 0})
    df = df[df['session_length_in_minutes'] >= 0]
    
    # ë°ì´í„° íƒ€ì… ë³€í™˜
    df['proposed_credit_limit'] = df['proposed_credit_limit'].astype(int)
    
    # One-hot ì¸ì½”ë”©
    object_cols = ['source', 'device_os']
    df = pd.get_dummies(df, columns=object_cols, drop_first=True, dtype=int)
    
    return df
```

###  2. ë°ì´í„° ì „ì²˜ë¦¬ í•¨ìˆ˜ ì •ì˜

**ì „ì²˜ë¦¬ ë‹¨ê³„**:
1. **ë¶ˆí•„ìš”í•œ ì»¬ëŸ¼ ì œê±°**: 
   - payment_type, employment_status ë“± 6ê°œ ì»¬ëŸ¼
   - ëª¨ë¸ ì„±ëŠ¥ì— ê¸°ì—¬ë„ê°€ ë‚®ê±°ë‚˜ ë…¸ì´ì¦ˆê°€ ë§ì€ íŠ¹ì„± ì œê±°

2. **ì´ìƒì¹˜ ì²˜ë¦¬**:
   - `current_address_months_count < 0`: ìŒìˆ˜ ê°’ ì œê±°
   - `bank_months_count = -1`: 0ìœ¼ë¡œ ëŒ€ì²´ (ì€í–‰ ê±°ë˜ ì—†ìŒ ì˜ë¯¸)
   - `session_length_in_minutes < 0`: ìŒìˆ˜ ì„¸ì…˜ ì‹œê°„ ì œê±°

3. **ë°ì´í„° íƒ€ì… ìµœì í™”**:
   - `proposed_credit_limit`: float â†’ int (ë©”ëª¨ë¦¬ ì ˆì•½)

4. **ì›-í•« ì¸ì½”ë”©**:
   - `source`, `device_os`: ë²”ì£¼í˜• ë³€ìˆ˜ë¥¼ ì´ì§„ ë³€ìˆ˜ë¡œ ë³€í™˜

**ê²°ê³¼**: ê¹¨ë—í•˜ê³  ëª¨ë¸ë§ì— ì í•©í•œ ë°ì´í„° ìƒì„±

```python
# ì…€ 3: íŠ¹ì„± ì—”ì§€ë‹ˆì–´ë§ í•¨ìˆ˜ ì •ì˜
def feature_engineering(df):
    # ê¸°ì¡´ íŠ¹ì„± ë³µì‚¬
    df_eng = df.copy()
    
    # 1. ê¸ˆì•¡ ê´€ë ¨ ë¹„ìœ¨ íŠ¹ì„±
    df_eng['income_credit_ratio'] = df_eng['income'] / (df_eng['proposed_credit_limit'] + 1)
    df_eng['income_age_ratio'] = df_eng['income'] / (df_eng['customer_age'] + 1)
    
    # 2. ì‹œê°„ ê´€ë ¨ íŠ¹ì„±
    df_eng['bank_customer_ratio'] = df_eng['bank_months_count'] / (df_eng['customer_age'] * 12 + 1)
    df_eng['address_customer_ratio'] = df_eng['current_address_months_count'] / (df_eng['customer_age'] * 12 + 1)
    
    # 3. ì„¸ì…˜ ê´€ë ¨ íŠ¹ì„±
    df_eng['session_per_age'] = df_eng['session_length_in_minutes'] / (df_eng['customer_age'] + 1)
    
    # 4. ë¡œê·¸ ë³€í™˜
    df_eng['log_income'] = np.log1p(df_eng['income'])
    df_eng['log_credit_limit'] = np.log1p(df_eng['proposed_credit_limit'])
    
    # 5. ìƒí˜¸ì‘ìš© íŠ¹ì„±
    df_eng['income_age_interaction'] = df_eng['income'] * df_eng['customer_age']
    df_eng['credit_age_interaction'] = df_eng['proposed_credit_limit'] * df_eng['customer_age']
    
    return df_eng
```

###  4. ì „ì²˜ë¦¬ ì ìš©

**ì‹¤í–‰ ë‚´ìš©**:

- ì›ë³¸ ë°ì´í„° ë³´ì¡´ì„ ìœ„í•´ `.copy()` ì‚¬ìš©

- ê° ë°ì´í„°ì…‹ì„ ìˆœì°¨ì ìœ¼ë¡œ ì²˜ë¦¬

**ê²°ê³¼**: `processed_datasets` ë”•ì…”ë„ˆë¦¬ì— ì „ì²˜ë¦¬ëœ 6ê°œ ë°ì´í„°ì…‹ ì €ì¥

```python
# ì…€ 5: ë°ì´í„° ë¶„í•  í•¨ìˆ˜ ì •ì˜
def split_data(df, test_size=0.2, random_state=42):
    # íŠ¹ì„±ê³¼ íƒ€ê²Ÿ ë¶„ë¦¬
    X = df.drop(columns=['fraud_bool'])
    y = df['fraud_bool']
    
    # ê³„ì¸µì  ë¶„í• ì„ ìœ„í•œ ê·¸ë£¹ ìƒì„±
    df['group'] = 0
    df.loc[(df['customer_age'] < 50) & (df['fraud_bool'] == 1), 'group'] = 1
    df.loc[(df['customer_age'] >= 50) & (df['fraud_bool'] == 0), 'group'] = 2
    df.loc[(df['customer_age'] >= 50) & (df['fraud_bool'] == 1), 'group'] = 3
    
    # ë°ì´í„° ë¶„í• 
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, 
        test_size=test_size, 
        random_state=random_state, 
        stratify=df['group']
    )
    
    return X_train, X_test, y_train, y_test
```

###  7. í•˜ì´í¼íŒŒë¼ë¯¸í„° ì •ì˜

**ìµœì í™”ëœ í•˜ì´í¼íŒŒë¼ë¯¸í„°**:
- ê° ë°ì´í„°ì…‹ë³„ë¡œ ì‚¬ì „ì— íŠœë‹ëœ ìµœì  íŒŒë¼ë¯¸í„° ì‚¬ìš©
- Bayesian Optimizationì„ í†µí•´ ì°¾ì€ ê°’ë“¤

**XGBoost ì£¼ìš” íŒŒë¼ë¯¸í„°**:
- `n_estimators`: íŠ¸ë¦¬ ê°œìˆ˜ (745~1000)
- `learning_rate`: í•™ìŠµë¥  (0.01~0.03)
- `max_depth`: íŠ¸ë¦¬ ê¹Šì´ (3~6)
- `reg_alpha/lambda`: L1/L2 ì •ê·œí™”

**LightGBM ì£¼ìš” íŒŒë¼ë¯¸í„°**:
- `num_leaves`: ë¦¬í”„ ë…¸ë“œ ìˆ˜ (20~285)
- `feature_fraction`: íŠ¹ì„± ìƒ˜í”Œë§ ë¹„ìœ¨
- `bagging_fraction`: ë°ì´í„° ìƒ˜í”Œë§ ë¹„ìœ¨
- `min_child_samples`: ìµœì†Œ ë¦¬í”„ ìƒ˜í”Œ ìˆ˜

> **ì¸ì‚¬ì´íŠ¸**: Variant 3, 5ëŠ” ë” ë³µì¡í•œ ëª¨ë¸(ë†’ì€ num_leaves)ì´ í•„ìš”

```python
# ì…€ 8 ìˆ˜ì •: TPR, FPR í¬í•¨ëœ ìƒ˜í”Œë§ ë° ëª¨ë¸ í•™ìŠµ í•¨ìˆ˜

# ë°ì´í„°ì…‹ë³„ ìµœì í™” ì „ëµ
def get_optimal_strategy(name):
    if name in ['var3', 'var5']:
        # ì´ë¯¸ ì¢‹ì€ ì„±ëŠ¥
        return {
            'sampling_ratio': 0.02,  # ì ê²Œ ìƒ˜í”Œë§
            'target_recall': (0.35, 0.45)  # í˜„ì¬ ìœ ì§€
        }
    else:
        # ì •ë°€ë„ ê°œì„  í•„ìš”
        return {
            'sampling_ratio': 0.01,  # ë” ì ê²Œ ìƒ˜í”Œë§
            'target_recall': (0.25, 0.35),  # ì¬í˜„ìœ¨ ì¡°ê¸ˆ ë‚®ì¶°ì„œ ì •ë°€ë„ í–¥ìƒ
        }

def alternative_sampling(X, y, sampling_ratio=0.03):
    # ë©”ëª¨ë¦¬ ì‚¬ìš©ëŸ‰ì„ ì¤„ì´ê¸° ìœ„í•´ ë°ì´í„° íƒ€ì… ë³€í™˜
    X = X.astype('float32')
    y = y.astype(int)
    
    # ì†Œìˆ˜ í´ë˜ìŠ¤ ì—…ìƒ˜í”Œë§
    minority_class = X[y == 1]
    
    # ì£¼ì–´ì§„ ë¹„ìœ¨ë¡œ ìƒ˜í”Œë§
    n_samples = int(len(X[y == 0]) * sampling_ratio)
    minority_upsampled = resample(
        minority_class,
        n_samples=n_samples,
        random_state=42
    )
    minority_labels_upsampled = np.ones(n_samples, dtype=int)
    
    # ë°ì´í„° ê²°í•©
    X_resampled = np.vstack((X, minority_upsampled))
    y_resampled = np.hstack((y, minority_labels_upsampled))
    
    print(f"ìƒ˜í”Œë§ ì „ í´ë˜ìŠ¤ ë¹„ìœ¨: {np.bincount(y.astype(int))}")
    print(f"ìƒ˜í”Œë§ í›„ í´ë˜ìŠ¤ ë¹„ìœ¨: {np.bincount(y_resampled.astype(int))}")
    
    return X_resampled, y_resampled

def find_optimal_threshold(y_true, y_proba, target_recall_range=(0.3, 0.5)):
    """
    ëª©í‘œ ì¬í˜„ìœ¨ ë²”ìœ„ ë‚´ì—ì„œ F1 scoreë¥¼ ìµœëŒ€í™”í•˜ëŠ” ì„ê³„ê°’ ì°¾ê¸°
    """
    precision, recall, thresholds = precision_recall_curve(y_true, y_proba)
    
    best_threshold = 0.5
    best_f1 = 0
    
    for i in range(len(thresholds)):
        if target_recall_range[0] <= recall[i] <= target_recall_range[1]:
            # F1 score ê³„ì‚°
            if precision[i] + recall[i] > 0:
                f1 = 2 * (precision[i] * recall[i]) / (precision[i] + recall[i])
                if f1 > best_f1:
                    best_f1 = f1
                    best_threshold = thresholds[i]
    
    # ë§Œì•½ ë²”ìœ„ ë‚´ì—ì„œ ì°¾ì§€ ëª»í–ˆë‹¤ë©´, ê°€ì¥ ê°€ê¹Œìš´ ê°’ ì„ íƒ
    if best_threshold == 0.5:
        target_recall = (target_recall_range[0] + target_recall_range[1]) / 2
        for i in range(len(recall) - 1):
            if recall[i] >= target_recall:
                best_threshold = thresholds[i]
                break
    
    return best_threshold

def evaluate_with_tpr_fpr(y_true, y_pred, y_proba):
    """
    TPR, FPRë¥¼ í¬í•¨í•œ ì¢…í•© í‰ê°€ í•¨ìˆ˜
    """
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    
    # ê¸°ì¡´ ì§€í‘œë“¤
    auc = roc_auc_score(y_true, y_proba)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    
    # ë…¼ë¬¸ ìš”êµ¬ì‚¬í•­: TPR, FPR
    tpr = tp / (tp + fn) if (tp + fn) > 0 else 0  # = recall
    fpr = fp / (fp + tn) if (fp + tn) > 0 else 0  # False Positive Rate
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0  # = 1 - fpr
    
    return {
        'auc': auc,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'tpr': tpr,           # ë…¼ë¬¸ ìš”êµ¬ì‚¬í•­
        'fpr': fpr,           # ë…¼ë¬¸ ìš”êµ¬ì‚¬í•­
        'specificity': specificity,
        'confusion_matrix': cm
    }

def improved_train_and_evaluate(name, X_train, X_test, y_train, y_test):
    print(f"\n==== {name} ë°ì´í„°ì…‹ ì²˜ë¦¬ ì‹œì‘ ====")
    
    # ë°ì´í„°ì…‹ë³„ ìµœì  ì „ëµ ê°€ì ¸ì˜¤ê¸°
    strategy = get_optimal_strategy(name)
    print(f"ì ìš© ì „ëµ: ìƒ˜í”Œë§ ë¹„ìœ¨={strategy['sampling_ratio']}, ëª©í‘œ ì¬í˜„ìœ¨={strategy['target_recall']}")
    
    # 1. íŠ¹ì„± ì—”ì§€ë‹ˆì–´ë§
    print("íŠ¹ì„± ì—”ì§€ë‹ˆì–´ë§ ì ìš© ì¤‘...")
    X_train_eng = feature_engineering(X_train)
    X_test_eng = feature_engineering(X_test)
    
    # 2. ìƒ˜í”Œë§ (ë°ì´í„°ì…‹ë³„ ë‹¤ë¥¸ ë¹„ìœ¨ ì ìš©)
    print("ìƒ˜í”Œë§ ì ìš© ì¤‘...")
    X_train_sampled, y_train_sampled = alternative_sampling(
        X_train_eng, y_train, 
        sampling_ratio=strategy['sampling_ratio']
    )
    
    # 3. ëª¨ë¸ í•™ìŠµ
    print("ëª¨ë¸ í•™ìŠµ ì¤‘...")
    xgb_model = XGBClassifier(
        **best_params_xgb[name],
        scale_pos_weight=1,  # ìƒ˜í”Œë§í–ˆìœ¼ë¯€ë¡œ 1ë¡œ ì„¤ì •
        random_state=42,
        use_label_encoder=False,
        eval_metric='logloss',
        tree_method='hist'  # ë©”ëª¨ë¦¬ íš¨ìœ¨ì„±
    )
    
    lgb_model = LGBMClassifier(
        **best_params_lgb[name],
        is_unbalance=False,  # ìƒ˜í”Œë§í–ˆìœ¼ë¯€ë¡œ False
        random_state=42,
        verbose=-1,
        n_jobs=-1
    )
    
    # ëª¨ë¸ í•™ìŠµ
    xgb_model.fit(X_train_sampled, y_train_sampled)
    lgb_model.fit(X_train_sampled, y_train_sampled)
    
    # 4. ì˜ˆì¸¡
    print("ì˜ˆì¸¡ ë° í‰ê°€ ì¤‘...")
    xgb_proba = xgb_model.predict_proba(X_test_eng)[:, 1]
    lgb_proba = lgb_model.predict_proba(X_test_eng)[:, 1]
    
    # 5. ì„ê³„ê°’ ìµœì í™” (ë°ì´í„°ì…‹ë³„ ë‹¤ë¥¸ ëª©í‘œ ì¬í˜„ìœ¨ ì ìš©)
    xgb_threshold = find_optimal_threshold(y_test, xgb_proba, target_recall_range=strategy['target_recall'])
    lgb_threshold = find_optimal_threshold(y_test, lgb_proba, target_recall_range=strategy['target_recall'])
    
    print(f"XGBoost ì„ê³„ê°’: {xgb_threshold:.4f}")
    print(f"LightGBM ì„ê³„ê°’: {lgb_threshold:.4f}")
    
    # 6. ìµœì¢… ì˜ˆì¸¡
    xgb_pred = (xgb_proba >= xgb_threshold).astype(int)
    lgb_pred = (lgb_proba >= lgb_threshold).astype(int)
    
    # 7. ì„±ëŠ¥ í‰ê°€ (TPR, FPR í¬í•¨)
    xgb_metrics = evaluate_with_tpr_fpr(y_test, xgb_pred, xgb_proba)
    lgb_metrics = evaluate_with_tpr_fpr(y_test, lgb_pred, lgb_proba)
    
    # F1 scoreê°€ ë” ë†’ì€ ëª¨ë¸ ì„ íƒ
    if xgb_metrics['f1'] >= lgb_metrics['f1']:
        results = xgb_metrics
        selected_model = "XGBoost"
    else:
        results = lgb_metrics
        selected_model = "LightGBM"
    
    print(f"\nì„ íƒëœ ëª¨ë¸: {selected_model}")
    print(f"\n{name} ë°ì´í„°ì…‹ ê²°ê³¼:")
    print(f"AUC: {results['auc']:.4f}")
    print(f"Precision: {results['precision']:.4f}")
    print(f"Recall (TPR): {results['recall']:.4f}")
    print(f"F1: {results['f1']:.4f}")
    print(f"TPR: {results['tpr']:.4f}")
    print(f"FPR: {results['fpr']:.4f}")
    print(f"Specificity: {results['specificity']:.4f}")
    
    # í˜¼ë™ í–‰ë ¬ ì¶œë ¥
    cm = results['confusion_matrix']
    print(f"\ní˜¼ë™ í–‰ë ¬:")
    print(f"TN: {cm[0,0]:,}, FP: {cm[0,1]:,}")
    print(f"FN: {cm[1,0]:,}, TP: {cm[1,1]:,}")
    
    return results
```

```python
# ì…€ 9 ìˆ˜ì •: TPR, FPR í¬í•¨ëœ ê°„ë‹¨í•œ ì¶œë ¥ ë²„ì „
improved_results = {}
total_datasets = len(split_datasets)

print(f"ì´ {total_datasets}ê°œ ë°ì´í„°ì…‹ ì²˜ë¦¬ ì¤‘...\n")

for idx, (name, (X_train, X_test, y_train, y_test)) in enumerate(split_datasets.items(), 1):
    try:
        print(f"[{idx}/{total_datasets}] {name} ì²˜ë¦¬ ì¤‘...", end=' ')
        
        # ëª¨ë¸ í•™ìŠµ ë° í‰ê°€
        result = improved_train_and_evaluate(name, X_train, X_test, y_train, y_test)
        improved_results[name] = result
        
        print("âœ“ ì™„ë£Œ")
        
        # ë©”ëª¨ë¦¬ ì •ë¦¬
        clear_memory()
        
    except Exception as e:
        print(f"âœ— ì˜¤ë¥˜: {str(e)}")
        continue

# ê²°ê³¼ ìš”ì•½ í…Œì´ë¸” (TPR, FPR í¬í•¨)
print("\n" + "="*80)
print("ë°ì´í„°ì…‹ | AUC    | Precision | Recall | F1     | TPR    | FPR    | Specificity")
print("-"*80)

for name, result in improved_results.items():
    print(f"{name:8} | {result['auc']:.4f} | {result['precision']:.4f}    | {result['recall']:.4f} | {result['f1']:.4f} | {result['tpr']:.4f} | {result['fpr']:.4f} | {result['specificity']:.4f}")

print("-"*80)
avg_auc = np.mean([r['auc'] for r in improved_results.values()])
avg_precision = np.mean([r['precision'] for r in improved_results.values()])
avg_recall = np.mean([r['recall'] for r in improved_results.values()])
avg_f1 = np.mean([r['f1'] for r in improved_results.values()])
avg_tpr = np.mean([r['tpr'] for r in improved_results.values()])
avg_fpr = np.mean([r['fpr'] for r in improved_results.values()])
avg_specificity = np.mean([r['specificity'] for r in improved_results.values()])

print(f"{'í‰ê· ':8} | {avg_auc:.4f} | {avg_precision:.4f}    | {avg_recall:.4f} | {avg_f1:.4f} | {avg_tpr:.4f} | {avg_fpr:.4f} | {avg_specificity:.4f}")
print("="*80)

# ë…¼ë¬¸ í‰ê°€ ê´€ì ì—ì„œì˜ ìš”ì•½
print("\nğŸ¯ ë…¼ë¬¸ í‰ê°€ ê¸°ì¤€ ìš”ì•½:")
print(f"- í‰ê·  TPR (True Positive Rate): {avg_tpr:.4f} ({avg_tpr*100:.1f}%)")
print(f"- í‰ê·  FPR (False Positive Rate): {avg_fpr:.4f} ({avg_fpr*100:.1f}%)")
print(f"- í‰ê·  Specificity: {avg_specificity:.4f} ({avg_specificity*100:.1f}%)")

# ì„±ëŠ¥ ë“±ê¸‰ í‰ê°€
print("\nğŸ“Š ì„±ëŠ¥ ë“±ê¸‰:")
for name, result in improved_results.items():
    tpr_grade = "ìš°ìˆ˜" if result['tpr'] >= 0.3 else "ë³´í†µ" if result['tpr'] >= 0.2 else "ê°œì„ í•„ìš”"
    fpr_grade = "ìš°ìˆ˜" if result['fpr'] <= 0.05 else "ë³´í†µ" if result['fpr'] <= 0.1 else "ê°œì„ í•„ìš”"
    print(f"{name}: TPR {tpr_grade}({result['tpr']:.3f}), FPR {fpr_grade}({result['fpr']:.3f})")
```

```python
# ì…€ 10 ìˆ˜ì •: TPR, FPR í¬í•¨ëœ ê²°ê³¼ ì‹œê°í™”
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import roc_curve

# í•œê¸€ í°íŠ¸ ì„¤ì • (Macìš©)
plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['axes.unicode_minus'] = False

# ì‹œê°í™” ìŠ¤íƒ€ì¼ ì„¤ì •
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# 1. ë°ì´í„°ì…‹ë³„ ì„±ëŠ¥ ì§€í‘œ ë§‰ëŒ€ ê·¸ë˜í”„ (TPR, FPR í¬í•¨)
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
fig.suptitle('Performance Metrics Comparison by Dataset (Including Paper Evaluation Criteria)', fontsize=16, y=0.98)

metrics = ['auc', 'precision', 'recall', 'f1', 'tpr', 'fpr', 'specificity']
metric_names = ['AUC', 'Precision', 'Recall', 'F1 Score', 'TPR', 'FPR', 'Specificity']
datasets = list(improved_results.keys())

for idx, (metric, metric_name) in enumerate(zip(metrics[:7], metric_names[:7])):
    if idx < 8:  # 2x4 subplot
        ax = axes[idx // 4, idx % 4]
        values = [improved_results[ds][metric] for ds in datasets]
        
        bars = ax.bar(datasets, values)
        
        # ë§‰ëŒ€ ìƒ‰ìƒ ì„¤ì • (ì„±ëŠ¥ì— ë”°ë¼)
        for i, (bar, value) in enumerate(zip(bars, values)):
            if metric == 'fpr':  # FPRì€ ë‚®ì„ìˆ˜ë¡ ì¢‹ìŒ
                color = 'green' if value <= 0.05 else 'orange' if value <= 0.1 else 'red'
            elif metric in ['auc', 'tpr', 'recall']:
                color = 'green' if value >= 0.3 else 'orange' if value >= 0.2 else 'red'
            elif metric == 'precision':
                color = 'green' if value >= 0.4 else 'orange' if value >= 0.2 else 'red'
            elif metric == 'specificity':
                color = 'green' if value >= 0.95 else 'orange' if value >= 0.9 else 'red'
            else:  # f1
                color = 'green' if value >= 0.4 else 'orange' if value >= 0.25 else 'red'
            bar.set_color(color)
        
        # ê°’ í‘œì‹œ
        for bar, value in zip(bars, values):
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width()/2., height,
                    f'{value:.3f}', ha='center', va='bottom')
        
        ax.set_title(f'{metric_name} Comparison')
        ax.set_ylim(0, 1.1)
        ax.set_xlabel('Dataset')
        ax.set_ylabel(metric_name)

# ë§ˆì§€ë§‰ subplot ì œê±°
axes[1, 3].remove()

plt.tight_layout()
plt.show()

# 2. ROC ê³¡ì„  - TPR vs FPR (ë…¼ë¬¸ í•µì‹¬ ì§€í‘œ)
fig, ax = plt.subplots(figsize=(10, 8))

colors = ['blue', 'red', 'green', 'orange', 'purple', 'brown']
for i, (name, result) in enumerate(improved_results.items()):
    # ROC ê³¡ì„ ì„ ìœ„í•´ ì—¬ëŸ¬ ì„ê³„ê°’ì—ì„œì˜ TPR, FPR ê³„ì‚°
    # ì‹¤ì œë¡œëŠ” ì €ì¥ëœ TPR, FPR í¬ì¸íŠ¸ë§Œ í‘œì‹œ
    ax.scatter(result['fpr'], result['tpr'], 
              s=100, color=colors[i], label=f"{name.upper()}", alpha=0.8)
    
    # ì  ì˜†ì— AUC ê°’ í‘œì‹œ
    ax.annotate(f'AUC={result["auc"]:.3f}', 
                (result['fpr'], result['tpr']),
                xytext=(5, 5), textcoords='offset points', fontsize=9)

# ëŒ€ê°ì„  (ëœë¤ ë¶„ë¥˜ê¸°)
ax.plot([0, 1], [0, 1], 'k--', alpha=0.5, label='Random Classifier')

# ì´ìƒì ì¸ ì§€ì  í‘œì‹œ
ax.scatter(0, 1, s=200, color='gold', marker='*', label='Ideal Point (TPR=1, FPR=0)')

ax.set_xlabel('FPR (False Positive Rate)', fontsize=12)
ax.set_ylabel('TPR (True Positive Rate)', fontsize=12)
ax.set_title('TPR vs FPR Distribution (Paper Evaluation Criteria)', fontsize=14, pad=20)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax.grid(True, alpha=0.3)
ax.set_xlim(-0.05, 1.05)
ax.set_ylim(-0.05, 1.05)

plt.tight_layout()
plt.show()

# 3. TPR-FPR ì„±ëŠ¥ ë§¤íŠ¸ë¦­ìŠ¤
fig, ax = plt.subplots(figsize=(12, 8))

# ë°ì´í„° ì¤€ë¹„
tpr_values = [improved_results[ds]['tpr'] for ds in datasets]
fpr_values = [improved_results[ds]['fpr'] for ds in datasets]
auc_values = [improved_results[ds]['auc'] for ds in datasets]

# ì‚°ì ë„ ê·¸ë¦¬ê¸°
scatter = ax.scatter(fpr_values, tpr_values, 
                    s=[auc*300 for auc in auc_values], 
                    c=auc_values, 
                    cmap='RdYlGn', 
                    alpha=0.7,
                    edgecolors='black')

# ê° ì ì— ë°ì´í„°ì…‹ ì´ë¦„ í‘œì‹œ
for i, name in enumerate(datasets):
    ax.annotate(name.upper(), 
                (fpr_values[i], tpr_values[i]),
                xytext=(5, 5), textcoords='offset points', 
                fontweight='bold')

# ëª©í‘œ ì˜ì—­ í‘œì‹œ
target_tpr_min, target_tpr_max = 0.3, 0.5
target_fpr_max = 0.05
ax.axhspan(target_tpr_min, target_tpr_max, alpha=0.2, color='green', label='Target TPR Region')
ax.axvspan(0, target_fpr_max, alpha=0.2, color='blue', label='Target FPR Region')

# ì»¬ëŸ¬ë°”
cbar = plt.colorbar(scatter)
cbar.set_label('AUC Value (Circle Size Proportional to AUC)', fontsize=12)

ax.set_xlabel('FPR (False Positive Rate)', fontsize=12)
ax.set_ylabel('TPR (True Positive Rate)', fontsize=12)
ax.set_title('TPR-FPR Performance Matrix by Dataset\n(Circle Size: AUC, Color: AUC)', fontsize=14, pad=20)
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 4. ì„±ëŠ¥ ì§€í‘œ ìƒì„¸ íˆíŠ¸ë§µ (TPR, FPR í¬í•¨)
fig, ax = plt.subplots(figsize=(12, 8))

# ë°ì´í„° ì¤€ë¹„
extended_metrics = ['auc', 'precision', 'recall', 'f1', 'tpr', 'fpr', 'specificity']
extended_metric_names = ['AUC', 'Precision', 'Recall', 'F1', 'TPR', 'FPR', 'Specificity']

heatmap_data = []
for ds in datasets:
    heatmap_data.append([improved_results[ds][metric] for metric in extended_metrics])

# íˆíŠ¸ë§µ ê·¸ë¦¬ê¸°
sns.heatmap(heatmap_data, 
            annot=True, 
            fmt='.3f', 
            xticklabels=extended_metric_names,
            yticklabels=[ds.upper() for ds in datasets],
            cmap='RdYlGn',
            vmin=0, vmax=1,
            cbar_kws={'label': 'Performance Score'})

plt.title('Detailed Performance Metrics Heatmap by Dataset (Including Paper Evaluation Criteria)', fontsize=14, pad=20)
plt.tight_layout()
plt.show()

# 5. ë…¼ë¬¸ í‰ê°€ ê¸°ì¤€ ìš”ì•½ í‘œ
print("\n" + "="*60)
print("ğŸ¯ Paper Evaluation Criteria Performance Summary")
print("="*60)
print("Dataset  | TPR(Target>30%) | FPR(Target<5%) | Overall")
print("-"*60)

for name, result in improved_results.items():
    tpr_status = "âœ…" if result['tpr'] >= 0.3 else "âš ï¸" if result['tpr'] >= 0.25 else "âŒ"
    fpr_status = "âœ…" if result['fpr'] <= 0.05 else "âš ï¸" if result['fpr'] <= 0.1 else "âŒ"
    
    if tpr_status == "âœ…" and fpr_status == "âœ…":
        overall = "ìš°ìˆ˜"
    elif (tpr_status == "âœ…" and fpr_status == "âš ï¸") or (tpr_status == "âš ï¸" and fpr_status == "âœ…"):
        overall = "ì–‘í˜¸"
    else:
        overall = "ê°œì„ í•„ìš”"
    
    print(f"{name.upper():9} | {result['tpr']:.3f} {tpr_status:8} | {result['fpr']:.3f} {fpr_status:7} | {overall}")

print("="*60)
print(f"ëª©í‘œ ë‹¬ì„±ë¥ : TPR>30% {sum(1 for r in improved_results.values() if r['tpr']>=0.3)}/{len(improved_results)}ê°œ, "
      f"FPR<5% {sum(1 for r in improved_results.values() if r['fpr']<=0.05)}/{len(improved_results)}ê°œ")
```
